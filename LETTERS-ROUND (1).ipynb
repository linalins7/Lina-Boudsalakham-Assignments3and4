```python
#LETTERS ROUND
from random import randint

words_list = [] #Global variable storing all reference words
L = [] #Global variable that will store letters for the player's game

def prompter():
    consonants = 'BCDFGHJKLMNPQRSTVWXYZ' #The program exclusively deals with uppercase characters
    vowels = 'AEIOU'
    for i in range(10): #Player is prompted ten times
        choice = input('"c" for consonant, "v" for vowel') 
        if choice == 'c':
            L.append(consonants[randint(0,20)]) #randint is used to select a random index
        else:
            L.append(vowels[randint(0,4)])
        print(L,end='\r') #We print multiple times but we only want to display one instance of the list at a time 
    print(L)

def word_finder(letters): #Takes in a list of letters and finds the longest word possible in 'words.txt'
       longest_word = ''
       result = True #We assume the word can be spelt with available letters until we find a letter that is not in the list.
       with open('words.txt','r') as file:
              for words in file:
                     words_list.append(words[:-1].upper()) #We use slicing it remove the '\n' suffix
       for word in words_list:
              result = True
              L = letters.copy() #We make copies of the list of available letters to avoid aliasing 
              for letter in word: #We verify if a word can be spelt with the available letters
                     if letter in L:
                            L.remove(letter) #Each available letter can only be used once
                     else:
                            result = False
              if result == True: #If the word can be spelt then we compare its length to longest_word 
                     if len(word) > len(longest_word): #If it is longer then we set longest_word to that word
                            longest_word = word
       return longest_word
     
def score(player_word,longest_word):
      return 100 - (len(longest_word) - len(player_word)) * 10
def letters_round():
    prompter()
    player_word = input('Write longest word with provided letters').upper()
    print(f'The longest word possible was {word_finder(L)}')
    print(f'Your score is {score(player_word,word_finder(L))}')
             
        
letters_round()
```

```python
#Numbers round
#Importing the module random so that we can work with random numbers given by the computer
import random

three_digits_target=random.randint(100,999) #We choose a random number of three digits

big_digits=[25,50,75,100]
final_digits=[]
for i in range(5): #In the number proposition given by the TV show, there are always 5 digits from the list of 1 to 10, and 1 number from the list of big numbers
    final_digits.append(random.randint(1,10)) #We choose 5 random numbers from 1 to 10
final_digits.append(random.choice(big_digits))#We choose 1 random number between 25,50,75,100

operators=['+','-','*','/']

result=0
numbers_used=[] #A list containing all the numbers that have already been used for operations
def players_guess():
    user_operation0=int(input("First number"))
    user_operation1=input("Give an operator")
    user_operation2=int(input("Last number"))
    operation=[user_operation0,user_operation1,user_operation2]
    operations_used=[] #A list containing all the results obtained before, so that we can use them as "First Number" or "Last number"
    for element in operation:
        if element in final_digits or element in operators or element in operations_used and element not in numbers_used:#We make sure the variables input in the beginning belong to the list of 6 random numbers, or the list of operants or also the list of results found before
            if user_operation1==operators[0]:
                result+=user_operation0 + user_operation2
            elif user_operation1==operators[1]:
                result+=user_operation0 - user_operation2
            elif user_operation1==operators[2]:
                result+=user_operation0 * user_operation2
            elif user_operation1==operators[3]:
                result+=user_operation0 / user_operation2
            operations_used.append(result)
    numbers_used.append(user_operation0) #We append the numbers that we just used because each number can be used once only
    numbers_used.append(user_operation2)
    return result

def difference_target():
    while result<three_digits_target and len(numbers_used)!=6: #we repeat the function players_guess as many times as we didn't get the target value or we didn't use all of the 6 numbers
        players_guess()
        diff=three_digits_target - result #difference between the target and the player's value
        return diff
             
def score():
    S=100-(three_digits_target-result)//10 #we return the score of the user
    return S
```
